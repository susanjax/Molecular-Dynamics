### Define box and water solvent
Now box to enclose the protein molecule is generated so that it will create a boundary for the simulation. This box is a cube which will put protein at its center, and later on water molecule will be added to fill the box.

```python
#create a box with parameter: a radius of 1nm, cube, protein at center
!gmx editconf -f prot_processed.gro -o prot_box.gro -c -d 1.0 -bt cubic
```

```python
#fill box with water
!gmx solvate -cp prot_box.gro -cs spc216.gro -o prot_solv.gro -p topol.top
#spc216.gro is a water model for three point water model (selected during water model), but we have choosed 4 point water model, and unsure this is the good point to use or not, further analysis will be carried out later
```

# Charge Balance by adding neutralizing ions
We know that there are -27 charge in the protein molecule, which you can find inside the topol.top file as the total charge of the protein molecule.

First download a ** ion.mdp ** file from 
http://www.mdtutorials.com/gmx/lysozyme/04_ions.html

This file is only intended to use with OPLS forcefield, if other force field is being used find new ions.mdp file or create one by yourself

```python
# create ions.tpr file
!gmx grompp -f ions.mdp -c prot_solv.gro -p topol.top -o ions.tpr
```

```python
#pass the file to genion
!printf "13" | gmx genion -s ions.tpr -o prot_solv_ions.gro -p topol.top -pname NA -nname CL -neutral
#replace the SOL (solvent) and add ions in their places; here SOL is in 13 position so it is choosen
#use this line in the command line and give input
#also keep visualizing using VMD
```

```python
!tail topol.top
#here we can see that total of 27 Na atoms had been replaced 
```

# Energy Minimization


```python
!gmx grompp -f minim.mdp -c prot_solv_ions.gro -p topol.top -o em.tpr
```

```python
!gmx mdrun -v -deffnm em
```

```python
#for visualization of energy minimization
!printf "10 0" | gmx energy -f em.edr -o potential.xvg
#the graph can be plotted using potential .xvg file and 
```

# Equilibrium 
There are two ways of conducting equilibrium,
First way is isothermal/isocratic in which NTV Number of particle, temperature,and volume are ke 

```python
!gmx grompp -f nvt.mdp -c em.gro -r em.gro -p topol.top -o nvt.tpr
```

```python
!gmx mdrun -deffnm nvt
#here the simulation will run for the 50000 steps = 100 ps
#nvt.mdp file can be modified to change the steps and time
```

```python
#check the temperature progression using energy
!printf "16 0" | gmx energy -f nvt.edr -o temperature.xvg 
```
